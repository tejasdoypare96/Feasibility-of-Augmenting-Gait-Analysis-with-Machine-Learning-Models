```python
import numpy as np

def find_transformation_matrix(image1, image2):
    # Step 1: Detect features and find matches
    orb = cv2.ORB_create()
    keypoints1, descriptors1 = orb.detectAndCompute(image1, None)
    keypoints2, descriptors2 = orb.detectAndCompute(image2, None)

    # Match features using BFMatcher
    bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    matches = bf.match(descriptors1, descriptors2)
    matches = sorted(matches, key=lambda x: x.distance)


    # Step 2: Extract matched keypoints
    source_points = np.float32([keypoints1[m.queryIdx].pt for m in matches]).reshape(-1, 2)
    target_points = np.float32([keypoints2[m.trainIdx].pt for m in matches]).reshape(-1, 2)
    # Step 3: Convert 2D points to homogeneous coordinates for 3D space
    source_points_h = np.hstack((source_points, np.zeros((source_points.shape[0], 1))))
    target_points_h = np.hstack((target_points, np.zeros((target_points.shape[0], 1))))


    # Step 4: Compute centroids
    src_centroid = np.mean(source_points_h, axis=0)
    trg_centroid = np.mean(target_points_h, axis=0)


    src_centered = source_points_h - src_centroid  # Center the points
    trg_centered = target_points_h - trg_centroid


    # Step 5: Compute the cross-covariance matrix
    CrossCovMat = np.dot(src_centered.T, trg_centered)



    # Step 6: Compute SVD
    U, S, Vt = np.linalg.svd(CrossCovMat)
    R = np.dot(Vt.T, U.T)


    if np.linalg.det(R) < 0:   # Ensure R is a proper rotation matrix (det(R) = 1)
        Vt[-1, :] *= -1
        R = np.dot(Vt.T, U.T)

    # Step 7: Compute the translation vector
    t = trg_centroid - np.dot(R, src_centroid)
    return R, t.reshape(-1, 1)


```

```python
if image1 is None or image2 is None:
  print("Error: Unable to load images.")
else:
  # Find the transformation matrices
  R, t = find_transformation_matrix(image1, image2)
  print("Rotation Matrix is:")
  print(R)
  print("\nTranslation Vector is:")
  print(t)
```
