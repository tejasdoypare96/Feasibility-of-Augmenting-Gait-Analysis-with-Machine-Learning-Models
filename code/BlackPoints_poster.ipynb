```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


def detect_black_points(image_path):
    # Load the input image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Crop the region of interest (ROI) where the poster is likely located
    height, width = gray.shape
    roi = gray[int(height * 0.2):int(height * 0.8), int(width * 0.2):int(width * 0.8)]

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(roi, (5, 5), 0)  # 5*5 kernal size

    # Threshold to isolate black points on the white poster
    _, binary = cv2.threshold(blurred, 60, 255, cv2.THRESH_BINARY_INV)

    # Detect contours from the binary image
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Prepare an empty list to store the coordinates of the detected white points
    coordinates = []

    # Iterate over all detected contours to find circular or square shapes
    for contour in contours:
        # Calculate the contour area to filter small noise points
        area = cv2.contourArea(contour)
        if 100 < area < 1000:  # Adjusted size constraints for poster points
            # Approximate the contour to check if it's circular or square
            perimeter = cv2.arcLength(contour, True)
            approx = cv2.approxPolyDP(contour, 0.02 * perimeter, True)
            if len(approx) >= 4:  # Looking for rectangular/square shapes
                # Get the centroid of the contour
                M = cv2.moments(contour)
                if M["m00"] != 0:
                    cX = int(M["m10"] / M["m00"]) + int(width * 0.2)
                    cY = int(M["m01"] / M["m00"]) + int(height * 0.2)
                    coordinates.append((cX, cY))

    # Sort the coordinates to ensure consistent raster-scan ordering
    coordinates.sort(key=lambda point: (point[1], point[0]))

    # Annotate points on the image with numbers from 1 to 15
    for idx, (x, y) in enumerate(coordinates[:15]):  # Limiting to 15 points
        cv2.circle(image, (x, y), 5, (0, 0, 255), -1)
        cv2.putText(image, f"{idx + 1}", (x - 10, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)



    # Save the processed image for verification
    output_path = image_path.replace(".png", "_processed.png")
    cv2.imwrite(output_path, image)

    # Print the coordinates in the required format
    for idx, (x, y) in enumerate(coordinates[:15]):
        print(f"Point {idx + 1}: ({x}, {y})")

    print(f"Processed image saved at: {output_path}")

    cv2_imshow(image)


```
