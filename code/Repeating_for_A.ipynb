```python
import traceback
import numpy as np
import json

try:
    # Define virtual coordinates in pixels (Q)
    Q = np.array([
        (0, 1072, 1375),
        (0, 1410, 1379),
        (0, 1754, 1382),
        (0, 1245, 1463),
        (0, 1582, 1468),
        (0, 1075, 1632),
        (0, 1245, 1635),
        (0, 1411, 1639),
        (0, 1578, 1640),
        (0, 1748, 1642),
        (0, 1244, 1802),
        (0, 1578, 1807),
        (0, 1075, 1886),
        (0, 1414, 1886),
        (0, 1742, 1894)
    ]).T

    # Convert virtual coordinates from pixels to cm
    Q = Q * 0.083

    # Define physical coordinates (P)
    P = np.array([
        (0, 154, 222),
        (0, 154, 194),
        (0, 154, 166),
        (0, 147, 208),
        (0, 147, 180),
        (0, 133, 222),
        (0, 133, 208),
        (0, 133, 194),
        (0, 133, 180),
        (0, 133, 166),
        (0, 119, 208),
        (0, 119, 180),
        (0, 112, 222),
        (0, 112, 194),
        (0, 112, 166)
    ]).T

    # Compute the mean of P and Q
    P_bar = np.mean(P, axis=1).reshape(-1, 1)
    Q_bar = np.mean(Q, axis=1).reshape(-1, 1)

    # Compute the covariance matrix
    P_cov = P - P_bar
    Q_cov = Q - Q_bar
    cov_mat = np.matmul(P_cov, Q_cov.T)

    # Perform SVD on the covariance matrix
    u, s, v = np.linalg.svd(cov_mat)

    # Compute the rotation matrix
    R = np.matmul(v.T, u.T)

    # Compute the translation vector
    T = Q_bar - np.matmul(R, P_bar)

    # Compute the error
    err = Q - (np.matmul(R, P) + np.array([T] * P.shape[1]).T)

    # Compute the inverse of the rotation matrix
    R_inv = np.linalg.inv(R)

    # Print the results as JSON
    print(json.dumps({
        "R": R.tolist(),
        "T": T.tolist(),
        "P": P.tolist(),
        "err": err.tolist(),
        "R_inv": R_inv.tolist()
    }))

except Exception as e:
    traceback.print_exc()
```
