```python
import yt_dlp
import os
import pandas as pd

# Define the path where videos will be saved
video_dir = 'videos/'
os.makedirs(video_dir, exist_ok=True)

# Function to download YouTube video using yt-dlp
def download_video(video_url, save_dir):
    ydl_opts = {
        'format': 'best',
        'outtmpl': os.path.join(save_dir, '%(title)s.%(ext)s'),  # Save as title of video
    }
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([video_url])
            video_info = ydl.extract_info(video_url, download=False)
            video_name = os.path.join(save_dir, video_info['title'] + ".mp4")
            return video_name
    except Exception as e:
        print(f"Error downloading {video_url}: {e}")
        return None

# Batch process the data, deduplicate URLs within batches, and download videos
def process_videos_in_batches(data, batch_size=500):
    # Keep track of downloaded and unavailable URLs globally across batches
    downloaded_urls = set()
    unavailable_urls = set()

    # Calculate total number of batches
    num_batches = len(data) // batch_size + (1 if len(data) % batch_size != 0 else 0)

    for batch_num in range(num_batches):
        # Get the current batch of data
        start_idx = batch_num * batch_size
        end_idx = min((batch_num + 1) * batch_size, len(data))
        batch_data = data[start_idx:end_idx]

        # Extract unique URLs in this batch
        batch_urls = batch_data['url'].apply(lambda x: x.strip()).unique()

        print(f"Processing batch {batch_num + 1}/{num_batches} with {len(batch_urls)} unique videos...")

        for video_url in batch_urls:
            # Skip if the URL is already marked as unavailable
            if video_url in unavailable_urls:
                print(f"Video {video_url} already marked as unavailable, skipping...")
                continue

            # Skip if the URL has already been downloaded
            if video_url in downloaded_urls:
                print(f"Video {video_url} already downloaded, skipping...")
                continue

            # Check if the video already exists in the folder
            video_name = f"{video_url.split('=')[-1]}.mp4"  # Use video ID as filename
            if os.path.exists(os.path.join(video_dir, video_name)):
                print(f"Video {video_name} already exists in the directory, skipping...")
                downloaded_urls.add(video_url)
                continue

            # Attempt to download the video
            video_path = download_video(video_url, video_dir)

            if video_path:
                print(f"Downloaded video at {video_path}")
                downloaded_urls.add(video_url)  # Mark the URL as downloaded
            else:
                print(f"Marking {video_url} as unavailable")
                unavailable_urls.add(video_url)  # Mark the URL as unavailable

# Example usage:
# Load your dataset (assuming it's a pandas DataFrame)
# data = pd.read_csv('your_file.csv')  # Replace with your actual CSV file path
# process_videos_in_batches(data)
```

```python
data = pd.read_csv('/content/GAVD_Clinical_Annotations_3.csv')
process_videos_in_batches(data)
```
