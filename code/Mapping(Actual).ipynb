## Virtual coordinates to physical coordinates

```python

# 3D Rotation matrix
R = np.array([[0.99901468, 0.04438102, 0.0],
              [0.04438102, -0.99901468, 0.0],
              [0.0, 0.0, 1.0]])

# 3D Translation vector
T = np.array([[44.62353032], [312.68703378], [0.0]])

# Virtual points
Q = np.array([
    (1223, 1745, 0), (1526, 1758, 0), (1842, 1768, 0),
    (1370, 1828, 0), (1679, 1842, 0), (1217, 1973, 0),
    (1366, 1985, 0), (1518, 1992, 0), (1673, 1998, 0),
    (1833, 2005, 0), (1362, 2139, 0), (1666, 2163, 0),
    (1213, 2205, 0), (1512, 2226, 0), (1823, 2245, 0)
])
Q = Q * 0.0923
Q = np.array(Q)

# Inverse the rotation matrix
R_inv = np.linalg.inv(R)

Q_translated = Q.T - T

# Apply the inverse rotation
P = (R_inv @ Q_translated).T
print("The physical coordinates in (cm):")
print(P)
```
